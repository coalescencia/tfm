```python
serie_meses = []

for x in range(1,13):
   
    mes_temporal = df[df['MES']== x]
    
    listado = (mes_temporal==0).sum(axis=1).tolist()
    sumatorio = 0
    
    for z in range(len(listado)):
        
        df_reducido = mes_temporal.iloc[:,7:]
        dias_nulos_mes = listado[z]
        #dias_nulos_mes = df_reducido.iloc[0].isnull().sum()
        if((lista_dia_meses[x-1]-dias_nulos_mes) == 0):
            sumatorio = sumatorio + df_reducido.iloc[0].sum()            
        else: 
            sumatorio = sumatorio + df_reducido.iloc[0].sum()/((lista_dia_meses[x-1])-dias_nulos_mes)
            
                
    serie_meses.append(sumatorio)
    
    
print(serie_meses)
    
```
