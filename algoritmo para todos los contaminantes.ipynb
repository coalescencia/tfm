<h3> PARTÍCULAS PM 10 </h3>

================================================================

```python
particulasPM10 = pd.DataFrame()
totales = []



    

for x in range(2001,2017):
    
    df = pd.read_excel('datos/PM_10/PM10_DD_'+str(x)+'.xlsx', sep=";")
    df_ordenado_por_meses= df.sort_values(by=['MES'])
    # Nos interesa únicamente las mediciones diarias por meses de todas las estaciones, para poder sacar 
    # el valor mensual del contaminante:
    df_datos_diarios_por_mes = df_ordenado_por_meses.iloc[:,6:]
   
    
    totales = []

    for p in range(1,13):
        
        medias_por_dia_mes = df_datos_diarios_por_mes[df_datos_diarios_por_mes['MES']== p ].iloc[:,1:].mean()
        media_por_mes = medias_por_dia_mes.mean()   
        totales.append(media_por_mes)
    serie1 = pd.Series(totales)
    
    particulasPM10['y'+str(x)] = serie1
```

```python
particulasPM10

```

```python
seriePm10 = pd.Series()

for j in range(2001,2017):
        
    seriePm10 = seriePm10.append(pd.Series(particulasPM10['y'+str(j)]))
    print(particulasPM10['y'+str(j)])
```
