```python
import pandas as pd
import numpy as np

import plotly.plotly as py
import plotly.graph_objs as go

import plotly
#plotly.tools.set_credentials_file(username='', api_key='')
```

```python
df = pd.read_csv('df_final.csv')
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'])
```

```python
df.set_index('Fecha')
```


<h3>Dashboard de visualización</h3>


Elaboración del dashboard con los datos completos de la evolución de la tasa de paro, así como la tasa de paro de los principales sectores de actividad Vs.contaminación registrada en el país por meses

```python
contaminante = go.Scatter(
                x=df.Fecha,
                y=df['contaminante'],
                name = "contaminante",
                line = dict(color = '#17BECF'),
                opacity = 0.8)

paro_total = go.Scatter(
                x=df.Fecha,
                y=df['paro'],
                name = "paro",
                line = dict(color = '#7F7F7F'),
                opacity = 0.8)

agricultura = go.Scatter(
                x=df.Fecha,
                y=df['Agricultura'],
                name = "Agricultura",
                line = dict(color = '#31B404'),
                opacity = 0.8)

manufacturera = go.Scatter(
                x=df.Fecha,
                y=df['Industria manufacturera'],
                name = "Industria manufacturera",
                line = dict(color = '#0000FF'),
                opacity = 0.8)


construccion = go.Scatter(
                x=df.Fecha,
                y=df['Construccion'],
                name = "Construccion",
                line = dict(color = '#e6e600'),
                opacity = 0.8)

comercio = go.Scatter(
                x=df.Fecha,
                y=df['Comercio'],
                name = "Comercio",
                line = dict(color = '#9900cc'),
                opacity = 0.8)

hosteleria = go.Scatter(
                x=df.Fecha,
                y=df['Hosteleria'],
                name = "Hosteleria",
                line = dict(color = '#ff8000'),
                opacity = 0.8)


administrativas = go.Scatter(
                x=df.Fecha,
                y=df['Actividades administrativas'],
                name = "Actividades administrativas",
                line = dict(color = '#669900'),
                opacity = 0.8)


admin_publicas = go.Scatter(
                x=df.Fecha,
                y=df['Administraciones publicas'],
                name = "Administraciones publicas",
                line = dict(color = '#0099ff'),
                opacity = 0.8)

sanitario = go.Scatter(
                x=df.Fecha,
                y=df['Sector sanitario'],
                name = "Sector sanitario",
                line = dict(color = '#ff99cc'),
                opacity = 0.8)



data = [contaminante,paro_total,agricultura,manufacturera,construccion,comercio, hosteleria,administrativas,admin_publicas,sanitario]

layout = dict(
    title = "Paro registrado total y por sectores vs. contaminación por PM10",
    xaxis = dict(
        range = ['2002-03-31','2016-12-31'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "manual")
```

**Nota:** Existe un pico de contaminación claramente visible en los meses de diciembre y de enero de 2016. Al tratarse de mediciones de 
partículas PM10 y como se explica en la memoria, los meses de invierno son muy propicios para una elevada tasa
de contaminación (se puede comprobar en el dashboard). Ese año fue especialmente seco y como confirman los datos 
del Ministerio se produjo una utilización masiva de medios de producción de electricidad procedentes de residuos 
fósiles y muy poca de energía renovable (principalmente hidroeléctrica)
