```python
completo = pd.read_csv('datos/completo_Contaminante_Paro.csv')
```

### Plotear todos las series temporales de paro total y contaminaci√≥n del dataset:

```python
ax = completo.plot(subplots=True,
                      layout=(2,2),
                      figsize=(20,16),
                      sharex=False,
                      sharey=False,
                      linewidth=1,
                      fontsize=5,
                      legend=False)

plt.show()
```

```python

ax = completo.plot(figsize= (20,14), colormap='Spectral', fontsize=6, linewidth=0.8)


ax.set_xlabel('Fecha', fontsize=10)
ax.set_ylabel('Tasa de desempleo', fontsize=10)
ax.set_title('Tasa de paro y contaminaci√≥n', fontsize=10)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))


#ax.axvline('2001-03-01', color='blue', linestyle='--', linewidth=0.8)
#ax.axvline('2016-12-31', color='blue', linestyle='--', linewidth=0.8)

plt.show()
```
